# Video Processing
1. Extract video frame size
2. know FPS
3. Save a video
4. Save a jpg file

```python
# read the video and convert save it to mp4 from second 14 to 30
import cv2
import matplotlib.pyplot as plt
```

```python
videopath = "../data/cityroller.mp4"
video = cv2.VideoCapture(videopath)
fps = video.get(cv2.CAP_PROP_FPS)
print("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))
```

```python
# get the width and height of the video
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
print("width: {}, height: {}".format(width, height))
```

```python
# Trim the video from second 14 to 30
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('../data/cityroller-trim.mp4', fourcc, fps, (width, height))
video.set(cv2.CAP_PROP_POS_FRAMES, fps*10)

# read the video and convert save it to mp4 from second 14 to 30
success, frame = video.read()
count = 0
for success, frame in iter(video.read, (False, None)):
    count += 1
    if count >= fps*14:
        out.write(frame)
    if count > fps*30:
        break
    
out.release()
```

```python
# plot the frame see how it looks
plt.imshow(frame)
# save one frame as jpg
cv2.imwrite("../data/cityroller-frame.jpg", frame)
```
